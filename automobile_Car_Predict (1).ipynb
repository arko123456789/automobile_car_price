# PROJECT PERPOSE 


The purpose of car price prediction is to estimate the value of a car in the market based on various factors, such as its make and model, year of manufacture, mileage, condition, and location. Car price prediction is useful for buyers who want to make an informed decision about purchasing a car, as well as for sellers who want to set a fair price for their vehicle. It can also be used by insurance companies, banks, and other financial institutions to determine the value of a car for insurance or loan purposes.

```python
# The train_test_split function of the sklearn. model_selection package in Python splits arrays or matrices into random subsets for train and test data, respectively.
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,
                                                   y,
                                                   test_size=0.20,
                                                   random_state=0)
```

```python
# Import LinearRegression allows us to fit a linear model to a dataset, predict new values, and evaluate the model's performance.
from sklearn.linear_model import LinearRegression
```

```python
# Using LinearRegression in New Dataset
automobile_Price=LinearRegression()
```

```python
#fit x_train & y_train in new_dataset
automobile_Price.fit(x_train,y_train)
```

```python
m = automobile_Price. coef_
n = automobile_Price.intercept_
m,n
```

```python
n
```

```python
# Predict x_train
y_predict_train = automobile_Price.predict(x_train)
y_predict_train
```

```python
# Predict x_test
y_predict_test = automobile_Price.predict(x_test)
y_predict_test
```

```python
# Import mean_absolute_error # The mean_absolute_error function creates two local variables, total and count that are used to compute the mean absolute error
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
# Checking mean_absolute_error between y_test & y_predict_test 
mean_absolute_error(y_test,y_predict_test)
```

```python
# Checking mean_absolute_error between y_train & y_predict_train
mean_absolute_error(y_train,y_predict_train)
```

```python
# Import mean_squared_error 
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# Checking mean_squared_error between y_test & y_predict_test
mean_squared_error(y_test,y_predict_test)
```

```python
# Checking mean_squared_error between y_train & y_predict_train
mean_squared_error(y_train,y_predict_train)
```

```python
# Checking accuracy between y_test & y_predict_test
accuracy=r2_score(y_test,y_predict_test)
```

```python
# Print aacuracy
accuracy
```

```python
# Print accuracy in percentage
print("accuracy is",accuracy*100)
```

```python
# Coefficient of determination between y_train & y_predict_train
r2score_train= r2_score(y_train, y_predict_train)
r2score_train
```

```python
# Using regplot for prediction price # regplot() method is used to plot data and draw a linear regression model fit.
plt.style.use("default")

plt.figure(figsize=(10, 10))
sns.regplot(x=y_test, y=y_predict_test, ci=None, color="black")
```

```python
# Using scatter plot between actual point & predict point # Scatter plots are used to plot data points on a horizontal and a vertical axis in the attempt to show how much one variable is affected by another
plt.figure(figsize=(10,10))
plt.style.use("ggplot")

plt.scatter(x=range(10),
            y=y_test.head(10),
            color="r",
           label="Actual Points")
plt.scatter(x=range(10),
            y=y_predict_test[:10],
            color="g",
           label="Predicted Points")
plt.legend()
```
